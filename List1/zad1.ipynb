```r
addition <- function(a, b){
    if (length(a) != length(b)) 0
    my_list <- c()
    for (i in 1:length(a)) {
        my_list[i] <- a[i] + b[i]
    }
    my_list
}
```

```r
scalar.mult <- function(a, b) {
    result = c()
    for (i in 1:length(b)) {
        result[i] <- a * b[i]
    }
    result
}
```

```r
dot.product <- function(a, b) {
    result = 0
    for (i in 1:length(a)) {
        result <- result + a[i] * b[i]
    }
    result
}
```

```r
matrix.addition <- function(a, b){
    long.c <- c()
    counter <- 1
    for (row in 1:nrow(a)) {
        for (col in 1:ncol(a)) {
            long.c[counter] <- a[row, col] + b[row, col]
            counter <- counter + 1
        }
    }
    matrix(long.c, nrow(a), ncol(a), TRUE)
}
```

```r
matrix.mul <- function(a, b){
    if (!ncol(a) == nrow(b)) {
        0
    } else {
        result <- matrix(0, nrow(a), ncol(b))

        for (row in 1:nrow(a)) {

            for (b_col in 1:ncol(b)) {

                 for (col in 1:ncol(a)) {
                     result[row, b_col] <- result[row, b_col] + (a[row, col] * b[col, b_col])
                }

            }

        }
        result
    }
}
```

```r
matrix.transpose <- function(a){
    result = matrix(0, ncol(a), nrow(a))
    for (col in 1:ncol(a)){
        result[col,] = a[,col]
    }
    result   
}
```

```r
m1 <- matrix(c(1,2,3,4,5,6), 3); m1
m2 <- matrix(c(1,2,3,4,5,6), 3)

x <- matrix.addition(m1, m2); x
x <- matrix.mul(m1, matrix.transpose(m2)); x
m1 %*% matrix.transpose(m2)
x <- matrix.transpose(m1); x
```

```r
#multiplies r-th row of M by factor
matrix.row.divide <- function( M, r, factor) {
    M[r,] = M[r,] * factor
    M #return
}
```

```r
#adds to i-th row tje j-th row multiplied by factor
matrix.add.raw <- function(i, j, factor, M) {
    M[i,] <- M[i,] + M[j,] * factor
    M # return
}
```

```r
# swap the i-th row with the j-th raw
matrix.swap.raw <- function(i, j, M) {
    tmp <- M[i,]
    M[i,] <- M[j,]
    M[j,] <- tmp
    M
}
```

```r
#general Gauss row reduction algorithm
# what if the matrix will be 2x10 dimension (it's ok for me no it's not) what if 10x2
matrix.eliminate <- function(m){
    border = min(c(nrow(m), ncol(m)))
    for (ic in 1:border){
        # find the index of max absolute value in column and swap rows
        to.swap <- which(m[ic:nrow(m),ic] == max(abs(m[ic:nrow(m),ic]))) + ic - 1
        m <- matrix.swap.raw(ic, to.swap[1], m)
        # makes one on the diagonal
        m <- matrix.row.divide(m, ic, 1/m[ic,ic]) #TODO fix the case m[1,1] == 0
        if (ic + 1 > nrow(m)) break
        for (ir in (ic+1):nrow(m)){
            m <- matrix.add.raw(ir, ic,-m[ir,ic], m)    
        }        
    }
    m
}
```

```r
matrix.solve <- function(m){
    # TODO check if invertible
    # append identity matrix to m
    mI <- cbind(m, diag(nrow(m)))
    mI <- matrix.eliminate(mI)
    for (i in nrow(mI):2) {
        for (j in (i-1):1) {
            mI <- matrix.add.raw(j, i, -mI[j,i], mI)    
        }
    }
    mI[,(nrow(m) + 1):ncol(mI)]
}
```

```r
matrix.rank <- function(m){
    border = min(c(nrow(m), ncol(m)))
    if (ncol(m) > nrow(m)) {
        m <- t(m)
    }
    rank <- border
    for (ic in 1:border){
        if (m[ic, ic] == 0){
            non.zero.index <- which(m[ic:nrow(m), ic] != 0) + ic -1
            # if exists non zero value in the column
            if(length(non.zero.index) > 0) {
                m <- matrix.swap.raw(non.zero.index[1], ic, m)
            } else {
                # there is no value different than zero
                # whole column is 0
                rank <- rank - 1
                next
            }
        }
        for (ir in 1:nrow(m)){
            if (ir != ic){
                m <- matrix.add.raw(ir, ic, -m[ir, ic]/m[ic,ic], m)
            }
        }
        #rank <- rank + 1
        # print(m)
    }
    rank
}
```
