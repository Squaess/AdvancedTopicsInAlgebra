```r
swap.rows <- function(i,j,M){
    tmp_row <- M[i,]
    M[i,] <- M[j,]
    M[j,] <- tmp_row
}

matrix.inverse.gauss <- function(a){
    concat <- cbind(a, diag(1, nrow(a), ncol(a)))
    for (i in 1:ncol(a)){
#         idx <- which.max(a[,i])
        a[i,] <- 1/a[i,i] * a[i,]
    }
    a
}
```
