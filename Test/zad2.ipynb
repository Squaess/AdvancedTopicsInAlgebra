```r
library(tidyverse)
```

```r
set.seed(2209760)
```

```r
A <- matrix( rnorm(5*5,mean=0,sd=1), 5, 5) 
A
```

```r
B <- t(A) %*% A
```

```r
B.eigen <- eigen(B)
B.eigen
```

```r
max.e <- max(B.eigen$values)
```

```r
max.e.v <- B.eigen$vectors[,which.max(B.eigen$values)]
```

```r
x.0 <- c(1.1,2,3.3,3,-5)
```

```r
L.0 <- max(Mod(x.0))
```

```r
x = x.0/L.0
```

```r
err <- rep(0, times=20)
```

```r
for (i in 1:20){
    p <- B %*% x
    L <- max(Mod(p))
    err[i] <- max.e - L
    x <- p/L
}
```

```r
max.e - L
```

```r
print(max.e.v)
print(x)
```
