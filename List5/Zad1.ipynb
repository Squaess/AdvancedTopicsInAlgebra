```r
# small dataset
set.seed(212)
Data = matrix(rnorm(30), 10, 3)

Data
```

```r
# centering with 'scale()'
center_scale <- function(x) {
    scale(x, scale = FALSE)
}

# apply it
Data <- center_scale(Data)
```

```r
pca.1 <- eigen(cov(Data))
pca.2 <- svd(Data)
print(pca.1)
print(pca.2)
```

```r
check_equal <-  function(M, v, e) {
    n <- nrow(M) - 1
    v^2/n
}

check_equal(Data, pca.2$d, pca.1$values)
```

```r
prco <- prcomp(Data)
print(prco$sdev^2)
print(prco$rotation)
```
