```r
library(purrr)
head(mtcars, 5)
mtcars %>%
  split(.$cyl) %>% # from base R
  map(~ lm(mpg ~ wt, data = .)) %>%
  map(summary) %>%
  map_dbl("r.squared")
```

```r
c1 <- rep(1:2, 4)
c2 <- rep(1, 8)
addition <- function(x, y){
    ret <- x+y
    print(x,y)
    ret
}
c1 %>%
    .^2 %>%
    sqrt(.) %>%
    +2 %>%
    map(addition)
```

```r
x <- c("why", "video", "cross", "extra", "deal", "authority")
str_length(x) 
#> [1] 3 5 5 5 4 9
str_c(x, collapse = ", ")
#> [1] "why, video, cross, extra, deal, authority"
str_sub(x, 1, 2)
#> [1] "wh" "vi" "cr" "ex" "de" "au"
```
