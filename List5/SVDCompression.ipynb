```r
library(imager)
```

```r
im <- load.image("image.jpg")
```

```r
plot(im)
```

```r
dim(im)
im.gray <- rowMeans(im, dims = 2)
```

```r
svd <- svd(im.gray)
```

```r
d <- svd$d
u <- svd$u
v <- svd$v
```

```r
composed = u %*% diag(d) %*% t(v)
plot(as.cimg(composed))
```

```r
composed = u[,1:10] %*% diag(d[1:10]) %*% t(v[,1:10])
plot(as.cimg(composed))
```

```r
for (i in c(3, 4, 5, 10, 20, 30))
{
	compressed <- u[,1:i] %*% diag(d[1:i]) %*% t(v[,1:i])
    plot(as.cimg(compressed))
}
```
